# Enumerate

Iterate with **enumerate** instead of **range(len(x))**

```python
data = [1, 2, -5, -2, 6]

for i in range(len(data)):
    if data[i] < 0:
        data[i] = 0

print(data)
```

```python
data = [1, 2, -5, -2, 6]

for idx, num in enumerate(data):
    if num < 0:
        data[idx] = 0
    
print(data)
```

# List Comprehension

Use **list comprehension** instead of raw for loops

```python
squares = []

for i in range(10):
    squares.append(i*i)
    
print(squares)
```

```python
squares = [i*i for i in range(10)]

print(squares)
```
